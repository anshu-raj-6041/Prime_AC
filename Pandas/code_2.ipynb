```python
import pandas as pd
```

#### Data Cleaning

```python
data = {
    "Name": ["Anshu", "Rahul", None, "Neha"],
    "Age": [21, None, 20, None],
    "Department": ["IT", "HR", "Finance", None],
    "Salary": [50000, 45000, None, 52000]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
df.dropna()     # any row having null value is removed
```

```python
df.dropna(how = "any")  # any row having null value is removed    
```

```python
df.dropna(how = "all")      # jb sbhi row NULL hoga tb remove hoga  
```

```python
df.fillna(0)    # fill null value with 0
```

```python
# replace it with mean value
df["Age"].fillna(df["Age"].mean())
```

```python
df["Salary"].fillna(df["Salary"].median())
```
