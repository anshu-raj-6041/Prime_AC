```python
l = []

for i in range(10):
    l.append(i)

print(l)
```

```python
range = np.arange(1,10,2)
range
```

```python
# np.linspace(start, stop, no of values b/w them)

arr = np.linspace(0,1,5)
arr
```

```python
# logarithmic scale array
# np.logspace(starting power, ending power, no of values b/w them)

arr = np.logspace(1, 3, 3)
arr
```

```python
# np.zeros => array full of zeros

arr = np.zeros(10)
arr
```

```python
arr = np.zeros([2,3])       # 2 rows, 3 columns
arr
```

```python
arr = np.ones(10)
arr
```

```python
arr = np.ones([2,3], int)       # we can also pass data type
arr
```

```python
# create an array full of any value
arr = np.full(10,2)
arr
```

```python
arr = np.full([2,4], 4.2)       # [row, column, default value]
arr
```

```python
# uninitialized array
arr = np.empty([2,3])
arr
```

```python
# generate random floats

arr = np.random.rand(2)
arr
```

```python
arr = np.random.randn(2,3)        # random floats from standard normal distribution
arr
```

```python
arr = np.random.randint(10, 100)        # random int
arr
```

```python
arr = np.random.randint(10, 100, size=[2,3])        # random int
arr
```
