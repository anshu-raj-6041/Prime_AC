#### Numpy Data Types and Type Casting 

```python
import numpy as np
```

```python
# numpy array does not allow multiple DT to store

arr = np.array([1,2,3,4,5])
arr
```

```python
print(type(arr))
arr.dtype
```

```python
arr = np.arange(10)
b = arr[1:5]
b[0] = 10

print(arr)
```

```python
arr = np.array([1,2.5,3,4])
arr.dtype
```

```python
arr = np.array([1,2,3,4], dtype=np.float32)
arr.dtype
```

```python
# type casting => astype()
```

```python
arr = np.array([1,2,3,4])
arr.dtype
```

```python
new_arr = arr.astype(np.float64)
new_arr.dtype
```
