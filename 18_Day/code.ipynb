```python
# Bar charts
import matplotlib.pyplot as plt

movies = [
    "3 idiots",
    "Border",
    "12 Fail",
    "KGF"
]

years = [2008, 2009, 2010, 2011]
revenue = [1005, 170, 427, 133]

plt.bar(years, revenue, color="green")
plt.title("Revenue in each year for movies")
plt.xlabel("Years")
plt.ylabel("Revenue ($M)")

plt.ylim(0, max(revenue)+100)
# plt.xlim(0, max(years)+5)

for i in range(len(years)):
    plt.text(years[i], revenue[i]+10, str(revenue[i]), ha="center", color="red")

plt.show()
```

```python
# compare 2 bar graph
import numpy as np


oscar_movies = [
    "3 idiots",
    "Border",
    "12 Fail",
    "KGF"
]
oscar_revenue = [1005, 170, 427, 133]

non_oscar_movies = [
    "Avatar",
    "Interstellar",
    "Balla",
    "Krish"
]
non_oscar_revenue = [125, 190, 227, 143]

years = [2008, 2009, 2010, 2011]

x = np.arange(len(years))
width = 0.4

plt.bar(x-(width/2), oscar_revenue, width, label="Oscar Movies")            # shifting to left
plt.bar(x+(width/2), non_oscar_revenue, width, label="Non-Oscar Movies")    # shifting to right   

plt.title("Oscar vs Non-Oscar movies revenue")
plt.xlabel("Years")
plt.ylabel("Revenue ($M)")

plt.legend()
plt.xticks(x, years)

```
